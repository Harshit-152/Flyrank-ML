## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

A fixed rule uses only a few conditions, such as low CTR or declining impressions.

Machine learning can combine many search and engagement signals to find patterns that are difficult to capture with simple rules.

The output helps prioritize pages for review and supports better decision-making.
```
