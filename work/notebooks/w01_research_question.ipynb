## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## Research Question

**Research question**
Which content pages should be prioritized for review based on their observed search and engagement signals?

**Decision**
Determine which pages should be reviewed first.

**Action**
A content manager reviews the highest-ranked pages and decides whether to update, rewrite, or optimize them.

**Output**
A ranked list of pages with a predicted priority score for review.

**Cost of a wrong recommendation**
If a page that truly needs attention is ranked too low, traffic and engagement may continue to decline. If a healthy page is ranked too high, time and resources may be wasted reviewing content that does not require updates.
```
